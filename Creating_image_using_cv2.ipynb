```python
# importing modules
import numpy as np
import cv2
```

```python
my_img = np.zeros((600, 600, 3), dtype = "uint8")
# creating Lines
cv2.line(my_img, (100, 170), (320, 20), (0, 20, 200), 10)
cv2.line(my_img, (320, 20), (500, 170), (0, 20, 200), 10)
# creating a rectangle
cv2.rectangle(my_img, (100, 170), (500, 400), (100, 20, 200), 10)
cv2.rectangle(my_img, (250, 300), (350, 400), (100, 20, 200), 10)
# Using cv2.putText() method
font = cv2.FONT_HERSHEY_SIMPLEX
cv2.putText(my_img, 'MY SWEET HOME', (210,520 ),font, 0.8, (255, 0, 0), 2, cv2.LINE_AA)
# Display image
cv2.imshow('Window', my_img)
cv2.waitKey(0)
cv2.destroyAllWindows()
```
