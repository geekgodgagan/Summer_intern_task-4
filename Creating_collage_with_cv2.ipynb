```python
# importing the modules
import cv2
import numpy as np
```

```python
# These are the two images
image1=cv2.imread("dravid.jpg")
image2=cv2.imread("Virat.jpg")
```

```python
image1.shape
```

```python
image2.shape
```

```python
# Croping images to make them of same size for fit
image1=cv2.resize(image1,(300,300))
image2=cv2.resize(image2,(300,300))
```

```python
# creating Horizontal stack
Horizontal=np.hstack([image1,image2])
```

```python
cv2.imshow("virat_dravid",Horizontal)
cv2.waitKey(0)
cv2.destroyAllWindows()
```
